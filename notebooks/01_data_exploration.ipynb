### Load All Datasets

The project analyzes three primary datasets:
1. **Enrolment Data** - New Aadhaar registrations
2. **Demographic Update Data** - People updating their address, name, DOB
3. **Biometric Update Data** - People updating their fingerprints/iris

```python
# Function to load all CSV files from a folder and combine them
def load_all_csvs(folder_path):
    """
    This function:
    1. Goes to the folder
    2. Finds all CSV files
    3. Loads them one by one
    4. Combines them into one big table
    """
    all_data = []
    all_data = []
    
    # Get list of all files in folder
    for filename in os.listdir(folder_path):
        if filename.endswith('.csv'):  # Only CSV files
            file_path = os.path.join(folder_path, filename)
            df = pd.read_csv(file_path)
            all_data.append(df)
            print(f"   Loaded: {filename} ({len(df):,} rows)")
    
    # Combine all into one table
    combined = pd.concat(all_data, ignore_index=True)
    return combined

print("Function created! Now let's load the data...")
```

```python
# Load Enrolment Data
print(" Loading Enrolment data...")
enrolment_df = load_all_csvs('../dataset/api_data_aadhar_enrolment')
print(f"\n Total Enrolment Records: {len(enrolment_df):,}\n")
```

```python
# Load Demographic Update Data
print(" Loading DEMOGRAPHIC UPDATE data...")
demographic_df = load_all_csvs('../dataset/api_data_aadhar_demographic')
print(f"\n Total Demographic Update Records: {len(demographic_df):,}\n")
```

```python
# Load Biometric Update Data
print(" Loading BIOMETRIC UPDATE data...")
biometric_df = load_all_csvs('../dataset/api_data_aadhar_biometric')
print(f"\n Total Biometric Update Records: {len(biometric_df):,}\n")
```

## Understanding Column Meanings

## Top 10 States by Enrolment

which states have the most Aadhaar enrolments.

```python
# Calculate total enrolments per state
enrolment_df['total_enrolled'] = enrolment_df['age_0_5'] + enrolment_df['age_5_17'] + enrolment_df['age_18_greater']

state_enrolment = enrolment_df.groupby('state')['total_enrolled'].sum().sort_values(ascending=False)

print(" TOP 10 STATES BY TOTAL ENROLMENT")
print("="*60)
for i, (state, count) in enumerate(state_enrolment.head(10).items(), 1):
    print(f"{i:2}. {state:25} : {count:>15,}")
```
